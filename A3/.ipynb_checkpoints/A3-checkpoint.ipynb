### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import glob
import tensorflow as tf
from random import randrange
from tensorflow.python import keras
from tensorflow.python.keras.models import Model
from tensorflow.python.keras.layers import Input, Conv2D, Conv2DTranspose, MaxPooling2D, concatenate, UpSampling2D
from tensorflow.python.keras.utils import Sequence
from tensorflow.python.keras.optimizers import Adam
from tensorflow.python.keras import backend as K
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

# Convolution operations

```python
def down(x, filters, kernel_size=(3, 3), padding="same", pool=True):
    conv = Conv2D(filters, kernel_size, padding=padding, activation="relu")(x)
    conv = Conv2D(filters, kernel_size, padding=padding, activation="relu")(conv)
    if pool:
        pool = MaxPooling2D()(conv)
        return conv, pool
    else:
        return conv

def up(x, residue, filters, kernel_size=(3, 3), padding="same"):
    upSample = Conv2D(filters, (2,2), padding=padding, activation="relu")(UpSampling2D()(x))
    concat = Concatenate()([residue, upSample])
    conv = Conv2D(filters, kernel_size, padding=padding, activation="relu")(concat)
    conv = Conv2D(filters, kernel_size, padding=padding, activation="relu")(conv)
    return conv
```

# Unet Model

```python
def Unet():
    f = [64, 128, 256, 512, 1024]
    inputs = Input((image_size, image_size, 3))
    
    img = inputs
    c1, p1 = down(img, f[0])
    c2, p2 = down(p1, f[1])
    c3, p3 = down(p2, f[2])
    c4, p4 = down(p3, f[3])
    
    mid = down(p4, f[4], pool=False)
    
    u1 = up(mid, c4, f[3])
    u2 = up(u1, c3, f[2])
    u3 = up(u2, c2, f[1])
    u4 = up(u3, c1, f[0])
    
    last = Conv2D(2, (3,3), padding="same", activation="relu")(u4)
    outputs = Conv2D(1, (1, 1), padding="same", activation="sigmoid")(last)
    unet = Model(inputs, outputs)
    return unet
```

# Managing inputs

```python
class InputSequence(Sequence):
    def __init__(self, inputlist, masklist, path, batch_size, image_size=128):
        self.inputlist = inputlist
        self.masklist = masklist
        self.path = path
        self.batch_size = batch_size
        self.image_size = image_size
        
    def __len__(self):
        return int(np.floor(len(self.inputlist) / self.batch_size))
    
    def __load__(self, img_name, mask_name):
        inputImg = cv2.imread(self.path + "input/" + img_name, 1)
        inputImg = cv2.resize(inputImg, (self.image_size, self.image_size))
        mask = cv2.imread(self.path + "mask/" + mask_name, cv2.IMREAD_GRAYSCALE)
        mask = cv2.resize(mask, (self.image_size, self.image_size))
        mask = np.expand_dims(mask, axis=-1)
        inputImg = inputImg/255.
        mask = mask/255.
        
        return inputImg, mask
    
    def __getitem__(self, index):
        if(index+1)*self.batch_size > len(self.inputlist):
            self.batch_size = len(self.inputlist) - index*self.batch_size
        
        input_images_batch = self.inputlist[index*self.batch_size : (index+1)*self.batch_size]
        mask_images_batch = self.masklist[index*self.batch_size : (index+1)*self.batch_size]
        img = []
        mask  = []
        
        for i in range(len(input_images_batch)):
            _img, _mask = self.__load__(input_images_batch[i], mask_images_batch[i])
            img.append(_img)
            mask.append(_mask)
            
        img = np.array(img)
        mask  = np.array(mask)
        return img, mask
```

### Loss Functions

```python
def dice_loss(y_true, y_pred, smooth=1):
    intersection = K.sum(y_true * y_pred, axis=[1,2,3])
    union = K.sum(y_true, axis=[1,2,3]) + K.sum(y_pred, axis=[1,2,3])
    return K.mean( (2. * intersection + smooth) / (union + smooth), axis=0)

def dice_coef(y_true, y_pred):
    return 1 - dice_loss(y_true, y_pred)

def bce(y_true, y_pred):
    return K.max(y_pred, 0) - y_pred * y_true + K.log(1 + K.exp(-K.abs(y_pred)))
    
def mse(y_true, y_pred):
    if (y_pred.shape[1]== None or y_pred.shape[2] == None):
        return K.sum((y_pred-y_true)**2) / (128*128)
    return K.sum((y_pred-y_true)**2) / (y_pred.shape[1]*y_pred.shape[2])
```

```python
def printTest(gen, model):
    fig=plt.figure(figsize=(20, 20))
    fig.subplots_adjust(hspace=0.4, wspace=0.4)
    columns = 5
    rows = 8
    for i in range(1, 21):
        x, y = gen.__getitem__(i-1)
        result = model.predict(x)
        result = result > 0.5
        fig.add_subplot(rows, columns, i)
        plt.imshow(np.reshape(y[0]*255, (image_size, image_size)), cmap="gray")
        plt.title("target {}".format(i))
        plt.axis('off')
        fig.add_subplot(rows, columns, i+20)
        plt.imshow(np.reshape(result[0]*255, (image_size, image_size)), cmap="gray")
        plt.title("mask {}".format(i))
        plt.axis('off')
    plt.savefig("q1_1.png")
    plt.show()
```

# Data Augmentation 

```python
images = [cv2.imread(file) for file in glob.glob(test_path+"input/*.jpg")]
masks = [cv2.imread(file) for file in glob.glob(test_path+"mask/*.jpg")]
```

```python
class AugmentInputSequence(InputSequence):
    def rotate(self, image, mask, angle):
        w = image.shape[1]
        h = image.shape[0]
        M = cv2.getRotationMatrix2D((w/2,h/2), angle, 1.0)
        image = cv2.warpAffine(image,M,(w,h))
        mask = cv2.warpAffine(mask,M,(w,h))
        mask = np.expand_dims(mask, axis=-1)
        return image, mask
    
    def height_shift(self, image, mask, height):
        w = image.shape[1]
        h = image.shape[0]
        M = np.float32([[1,0,0],[0,1,height]])
        image = cv2.warpAffine(image,M,(w,h))
        mask = cv2.warpAffine(mask,M,(w,h))
        mask = np.expand_dims(mask, axis=-1)
        return image, mask
    
    def width_shift(self, image, mask, width):
        w = image.shape[1]
        h = image.shape[0]
        M = np.float32([[1,0,width],[0,1,0]])
        image = cv2.warpAffine(image,M,(w,h))
        mask = cv2.warpAffine(mask,M,(w,h))
        mask = np.expand_dims(mask, axis=-1)
        return image, mask
    
    def flip(self, image, mask):
        imageResults = [image]
        maskResults = [mask]
        for i in [-1,0,1]:
            img = cv2.flip(image, flipCode=i)
            msk = cv2.flip(mask, flipCode=i)
            msk = np.expand_dims(msk, axis=-1)
            imageResults.append(img)
            maskResults.append(msk)
        return imageResults, maskResults
    
    def augment(self, image, mask, copies):
        imgList, maskList = self.flip(image, mask)
        imageResults = []
        maskResults = []
        for i in range(len(imgList)):
            imageResults.append(imgList[i])
            maskResults.append(maskList[i])
            w = image.shape[1]
            h = image.shape[0]
            for j in range(copies):
                randAngle = randrange(0, 180)
                randHeight = randrange(-h/2, h/2)
                randWidth = randrange(-w/2, w/2)
                rotimg, rotmask = self.rotate(imgList[i], maskList[i], randAngle)
                imageResults.append(rotimg)
                maskResults.append(rotmask)
                hsimg, hsmask = self.height_shift(imgList[i], maskList[i], randHeight)
                imageResults.append(hsimg)
                maskResults.append(hsmask)
                wsimg, wsmask = self.width_shift(imgList[i], maskList[i], randWidth)
                imageResults.append(wsimg)
                maskResults.append(wsmask)
        return imageResults, maskResults
    
    def __getitem__(self, index):
        if(index+1)*self.batch_size > len(self.inputlist):
            self.batch_size = len(self.inputlist) - index*self.batch_size
        
        input_images_batch = self.inputlist[index*self.batch_size : (index+1)*self.batch_size]
        mask_images_batch = self.masklist[index*self.batch_size : (index+1)*self.batch_size]
        img = []
        mask  = []
        
        for i in range(len(input_images_batch)):
            _img, _mask = self.__load__(input_images_batch[i], mask_images_batch[i])
            _img, _mask = self.augment(_img, _mask, 4)
            for j in range(len(_img)):
                img.append(_img[j])
                mask.append(_mask[j])
            
        img = np.array(img)
        mask  = np.array(mask)
        
        
        return img, mask
```
