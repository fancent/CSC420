# Question 1

```python
beeImg = cv2.imread("bee.jpg")
b,g,r = cv2.split(beeImg)

def interpolation1D(image, size):
    h = [i/size for i in range(size)] + [1] + [i/size for i in range(size-1,-1, -1)]
    result = []
    for i in range(len(image)):
        #padded = [0 for i in range((size-1)*(len(image[0])-1)+len(image[0]))]
        padded = [0. for i in range((len(image[0])-1)*size +1)]
        padded[::size] = image[i]
        padded = np.pad(padded, (size), 'constant')
        temp= []
        for j in range(size, len(padded)-size):
            sliced = padded[(j-size):(j+size+1)]
            product = np.matmul(h, sliced)
            temp.append(product)
        result.append(temp)
    return np.asarray(result)
```

```python
bx4 = interpolation1D(b, 4)
gx4 = interpolation1D(g, 4)
rx4 = interpolation1D(r, 4)

by4 = interpolation1D(bx4.T, 4)
gy4 = interpolation1D(gx4.T, 4)
ry4 = interpolation1D(rx4.T, 4)

merged = cv2.merge((by4.T, gy4.T, ry4.T))
cv2.imwrite('Q1a.png', merged)
```

```python
def interpolation2D(image, size):
    h = [i/size for i in range(size)] + [1] + [i/size for i in range(size-1,-1, -1)]
    hT = [[i/size] for i in range(size)] + [[1]] + [[i/size] for i in range(size-1,-1, -1)]
    kernal = np.multiply(hT, h)
    
    padded = []
    for i in range(len(image)):
        temp = [0. for i in range((len(image[0])-1)*size +1)]
        temp[::size] = image[i]
        padded.append(temp)
        if i < len(image)-1:
            for j in range(size-1):
                temp2 = [0. for i in range((len(image[0])-1)*size +1)]
                padded.append(temp2)
    
    result = signal.convolve2d(padded, kernal, mode='same')
    return result
```

```python
b = interpolation2D(b, 4)
g = interpolation2D(g, 4)
r = interpolation2D(r, 4)

merged = cv2.merge((b, g, r))
cv2.imwrite('Q1b.png', merged)
```

# Question 2

##### Part A

```python
img = cv2.imread('building.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
def Harris(image, alpha):
    newImg = image.copy()
    blur = cv2.GaussianBlur(image,(5,5),7)
    Ix = cv2.Sobel(blur, cv2.CV_64F, 1, 0, ksize=5)
    Iy = cv2.Sobel(blur, cv2.CV_64F, 0, 1, ksize=5)
    
    IxIy = np.multiply(Ix, Iy)
    Ix2 = np.multiply(Ix, Ix)
    Iy2 = np.multiply(Iy, Iy)
    
    Ix2_blur = cv2.GaussianBlur(Ix2,(7,7),10) 
    Iy2_blur = cv2.GaussianBlur(Iy2,(7,7),10) 
    IxIy_blur = cv2.GaussianBlur(IxIy,(7,7),10)
    
    det = np.multiply(Ix2_blur, Iy2_blur) - np.multiply(IxIy_blur,IxIy_blur)
    trace = Ix2_blur + Iy2_blur
    R = det - alpha * np.multiply(trace,trace)
    
    rows, cols = np.where(R>0.01*np.max(R))
    for i, j in zip(rows,cols):
        newImg = cv2.circle(newImg, (j,i), radius=3, color=0,thickness=0)
    return newImg
```

```python
#Harris
R_H = Harris(gray, 0.04)
fig1, graph = plt.subplots(figsize=(10,5))
graph.imshow(R_H,cmap = 'gray')
graph.axis('off')
cv2.imwrite('Q2aHarris.png', R_H)
```

```python
def Brown(image, alpha):
    newImg = image.copy()
    blur = cv2.GaussianBlur(image,(5,5),7)
    Ix = cv2.Sobel(blur, cv2.CV_64F, 1, 0, ksize=5)
    Iy = cv2.Sobel(blur, cv2.CV_64F, 0, 1, ksize=5)
    
    IxIy = np.multiply(Ix, Iy)
    Ix2 = np.multiply(Ix, Ix)
    Iy2 = np.multiply(Iy, Iy)
    
    Ix2_blur = cv2.GaussianBlur(Ix2,(7,7),10) 
    Iy2_blur = cv2.GaussianBlur(Iy2,(7,7),10) 
    IxIy_blur = cv2.GaussianBlur(IxIy,(7,7),10)
    
    det = np.multiply(Ix2_blur, Iy2_blur) - np.multiply(IxIy_blur,IxIy_blur)
    trace = Ix2_blur + Iy2_blur
    R = det / (trace+alpha)
    
    rows, cols = np.where(R>0.05*np.max(R))
    for i, j in zip(rows,cols):
        newImg = cv2.circle(newImg, (j,i), radius=3, color=0,thickness=0)
    return newImg
```

```python
#Brown
R_B = Brown(gray,0.04)
fig2, graph =plt.subplots(figsize=(10,5))
graph.imshow(R_B,cmap = 'gray')
graph.axis('off')
cv2.imwrite('Q2aBrown.png', R_B)
```

Brown's method takes the harmonic mean which has a smoother change than computing difference of determinant and trace.
It is especially smooth in region where Lambda1 ~ Lambda2

##### Part C

```python
def laplaceOfGauss(image, scale):
    size = scale*2 - 1
    blurred = cv2.GaussianBlur(image, (size,size), 0)
    result = cv2.Laplacian(blurred, cv2.CV_64F, ksize=size)
    minimum = np.min(result)
    return (result < minimum*0.6) * result

def logscale(image, minScale, maxScale):
    result = image.copy()
    for s in range(minScale, maxScale):
        temp = image.copy()
        mask = laplaceOfGauss(temp, s)
        rows, cols = np.where(mask < 0)
        for i, j in zip(rows,cols):
            result = cv2.circle(result, (j,i), radius=5+s, color=0,thickness=0)
    return result
```

```python
R_LoG = logscale(gray, 3,12)
#R_LoG = laplaceOfGauss(gray, 9)
fig2, graph =plt.subplots(figsize=(20,10))
graph.imshow(R_LoG,cmap = 'gray')
graph.axis('off')
cv2.imwrite('Q2c.png', R_LoG)
```

#### Part B

```python
def LNorm(pt1, pt2, power):
    result = 0
    for i in range(len(pt1)):
        result += np.abs(pt1[i] - pt2[i])**power
    return result ** (1./power)

def find_matches(set1, set2, power, threshold):
    kpResult = []
    for i in range(len(set1)):
        temp = []
        for j in range(len(set2)):
            temp.append([LNorm(set1[i], set2[j], power),j])
        temp.sort()
        firstClosest = temp[0]
        secondClosest = temp[1]
        if firstClosest[0]/secondClosest[0] < threshold:
            kpResult.append([cv2.DMatch(i,firstClosest[1], firstClosest[0])])
    return kpResult
```

##### Creating the matched line image

```python
normResults = find_matches(des1, des2, 2, 0.6)
```

```python
def removeSimilarMatches(m1, m2):
    return np.abs(m1[0].queryIdx-m2[0].queryIdx) > 15

def sortTop10(matches):
    sortedMatches = sorted(matches,key=lambda x: x[0].distance)
    result = []
    for element in sortedMatches:
        if all(removeSimilarMatches(element, others) for others in result):
            result.append(element)
    return result[:10]
```

```python
sortResults = sortTop10(normResults)
matched_img = cv2.drawMatchesKnn(sample1,kp1,sample2,kp2, sortResults, None, flags=2)
cv2.imwrite('Q4b.png', matched_img)
```

##### Creating the graph 

```python
thresholdRange = np.arange(0.2,0.8,0.05)
numOfMatchesL2 = [len(find_matches(des1, des2, 2, t)) for t in thresholdRange]
```

```python
fig, graph = plt.subplots()
graph.plot(thresholdRange, numOfMatchesL2)
graph.set(xlabel="Threshold Range", ylabel="number of matches",
       title="Threshold Range over of matches by using L2-norm")
graph.grid()
fig.savefig("Q4b_Graph.png")
plt.show()
```

```python
def addGaussNoise(image, mean, sigma):
    noise = np.random.normal(mean, sigma, image.shape)
    result = image+noise
    return np.clip(result,0,1)
```

#### Part E

```python
colorTemplate = cv2.imread('colourTemplate.png')
colorSearch = cv2.imread('colourSearch.png')
b1,g1,r1 = cv2.split(colorTemplate)
b2,g2,r2 = cv2.split(colorSearch)
```

```python
cv2.imwrite('b1.png',b1)
cv2.imwrite('b2.png',b2)
cv2.imwrite('g1.png',g1)
cv2.imwrite('g2.png',g2)
cv2.imwrite('r1.png',r1)
cv2.imwrite('r2.png',r2)
```

```python
def find_matches_rgb(rgb1, rgb2,key_pts1, key_pts2, des_set1, des_set2, power, threshold):
    kpResult = []
    for i in range(len(des_set1)):
        temp = []
        for j in range(len(des_set2)):
            temp.append([LNorm(des_set1[i], des_set2[j], power),j])
        temp.sort()
        firstClosest = temp[0]
        secondClosest = temp[1]
        if firstClosest[0] < threshold*secondClosest[0]:
            x1 , y1 = key_pts1[i].pt
            x2 , y2 = key_pts2[firstClosest[1]].pt
            if rgb1[int(y1)][int(x1)].all() == rgb2[int(y2)][int(x2)].all():
                kpResult.append([cv2.DMatch(i,firstClosest[1], firstClosest[0])])
    return kpResult
```

```python
sift = cv2.xfeatures2d.SIFT_create()
kp_b1, des_b1 = sift.detectAndCompute(b1,None)
```

```python
sift = cv2.xfeatures2d.SIFT_create()
kp_b2, des_b2 = sift.detectAndCompute(b2,None)
```

```python
sift = cv2.xfeatures2d.SIFT_create()
kp_r1, des_r1 = sift.detectAndCompute(r1,None)
```

```python
sift = cv2.xfeatures2d.SIFT_create()
kp_r2, des_r2 = sift.detectAndCompute(r2,None)
```

```python
sift = cv2.xfeatures2d.SIFT_create()
kp_g1, des_g1 = sift.detectAndCompute(g1,None)
```

```python
sift = cv2.xfeatures2d.SIFT_create()
kp_g2, des_g2 = sift.detectAndCompute(g2,None)
```

```python
total_des_1 = np.concatenate((des_b1,des_r1),axis=0)
total_des_2 = np.concatenate((des_b2,des_r2),axis=0)
kp_1 = kp_b1+kp_r1
kp_2 = kp_b2+kp_r2
```

```python
normResults = find_matches_rgb(colorTemplate, colorSearch,kp_1,kp_2,total_des_1,total_des_2, 2, 0.4)
```

```python
matched_img = cv2.drawMatchesKnn(colorTemplate,kp_b1 + kp_r1,colorSearch,kp_b2 + kp_r2, normResults, None, flags=2)
cv2.imwrite('Q4e_match.png', matched_img)
```
